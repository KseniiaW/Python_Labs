```python
from random import randint
import time

igrok1 = input('Введите имя 1-го играющего: ')
igrok2 = input('Введите имя 2-го играющего: ')

sum1 = 0  #сумма очков первого игрока
sum2 = 0  # сумма очков второго игрока
wins1 = 0  # количество побед первого игрока
wins2 = 0  # количество побед второго игрока

print(f'\n{igrok1} vs {igrok2} - играем 5 партий!\n')

# Цикл для 5 партий
for i in range(1, 6):
    print(f'=== Партия {i} ===')
    
    # Бросок первого игрока
    print(f'{igrok1} бросает кубик...', end=' ')
    time.sleep(1.5)
    n1 = randint(1, 6)
    print(f'Выпало: {n1} ({"⚀⚁⚂⚃⚄⚅"[n1-1]})')
    
    # Бросок второго игрока
    print(f'{igrok2} бросает кубик...', end=' ')
    time.sleep(1.5)
    n2 = randint(1, 6)
    print(f'Выпало: {n2} ({"⚀⚁⚂⚃⚄⚅"[n2-1]})')
    
    # Обновление счетчиков
    sum1 += n1
    sum2 += n2
    
    # Определение победителя партии
    if n1 > n2:
        wins1 += 1
        print(f'Партию выиграл {igrok1}\n')
    elif n1 < n2:
        wins2 += 1
        print(f'Партию выиграл {igrok2}\n')
    else:
        print('Ничья в этой партии!\n')
    
    time.sleep(1)

# Итоги игры
print('\n=== ИТОГИ ИГРЫ ===')
print(f'Сумма очков: {igrok1} - {sum1}, {igrok2} - {sum2}')
print(f'Побед: {igrok1} - {wins1}, {igrok2} - {wins2}')

# Определение победителя
if sum1 > sum2:
    print(f'\nПОБЕДИТЕЛЬ: {igrok1} по сумме очков!')
elif sum1 < sum2:
    print(f'\nПОБЕДИТЕЛЬ: {igrok2} по сумме очков!')
else:
    if wins1 > wins2:
        print(f'\nПОБЕДИТЕЛЬ: {igrok1} по количеству побед!')
    elif wins1 < wins2:
        print(f'\nПОБЕДИТЕЛЬ: {igrok2} по количеству побед!')
    else:
        print('\nАБСОЛЮТНАЯ НИЧЬЯ!')

print('\nИгра окончена!')
```
